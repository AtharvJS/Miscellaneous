```python
# Creating a function for creating a dictionary of lists of length of scrape_list
def create_category_lists(num_categories):
  """Creates a dictionary of empty lists with keys in the format 'category_i'.

  Args:
    num_categories: The number of categories (lists) to create.

  Returns:
    A dictionary with keys as 'category_i' and values as empty lists.
  """

  category_dict = {}
  for i in range(1, num_categories + 1):
    category_dict[f"category_{i}"] = []
  return category_dict
```

# Function

```python
# Category scraping function
def scraper(html):
  titles = []
  prices = []
  sale_of_past_months = []
  ratings = []
  no_of_ratings = []

  title = html.find('span', id = 'productTitle').text.strip()
  titles.append(title)

  price = html.find('span', class_='a-price-whole').text.strip()
  prices.append(price)

  sale_past_month = html.find('span', class_='a-size-small social-proofing-faceout-title-text').text.strip()
  sale_of_past_months.append(sale_past_month)

  rating = html.find('span', class_='a-size-base a-color-base').text.strip()
  ratings.append(rating)

  no_of_rating = html.find('span', id='acrCustomerReviewText').text.strip()
  no_of_ratings.append(no_of_rating)
  return titles, prices, sale_of_past_months, ratings, no_of_ratings

```
